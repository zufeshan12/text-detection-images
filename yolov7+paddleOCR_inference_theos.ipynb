```python
import sys
home_dir = '/content/text-detection-images/easy-yolov7/'
sys.path.append(home_dir)
from algorithm.object_detector import YOLOv7
from utils.detections import draw
import json
import cv2
import os
```

```python
'''if not(os.path.exists(home_dir+'images-with-text/')):
    os.mkdir(home_dir+'images-with-text')
if not(os.path.exists(home_dir+'images-with-text/detections')):
    os.mkdir(home_dir+'images-with-text/detections')'''
```

```python
os.makedirs(home_dir+'images-with-text',exist_ok=True)
os.makedirs(home_dir+'images-with-text/detections',exist_ok=True)
```

```python
import glob
import sys
sys.argv=['']
del sys

yolov7 = YOLOv7()
yolov7.set(ocr_classes=['text'])
yolov7.load(home_dir+'best.weights'
            ,classes=home_dir+'classes.yaml'
            ,ocr_weights=home_dir+'best', device='cpu') # use 'gpu' for CUDA GPU inference
for img in glob.glob(home_dir+'images-with-text/*.JPG'):
    img_name = img.split('/')[-1]
    image = cv2.imread(img)
    detections = yolov7.detect(image)
    detected_image = draw(image, detections)
    cv2.imwrite(home_dir+'images-with-text/detections/detected_'+img_name, detected_image)
    #print(json.dumps(detections, indent=4))

```
