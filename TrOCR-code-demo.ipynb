```python
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import requests
from PIL import Image
from IPython.display import display
```

## Implementation

### For handwritten texts

```python
#for handwritten texts
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-large-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-large-handwritten")
```

```python
image = Image.open("/content/S1013891.JPG").convert("RGB")
display(image)
```

```python
pixel_values = processor(image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
display(processor.batch_decode(generated_ids, skip_special_tokens=True)[0])
```

### For printed images and scanned documents

```python
#for printed images and scans
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-printed")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-printed")
```

```python
# load image from the IAM dataset
#url = "https://fki.tic.heia-fr.ch/static/img/a01-122-02.jpg"
#image = Image.open(requests.get(url, stream=True).raw).convert("RGB")
image = Image.open("/content/S1013769.JPG").convert("RGB")
pixel_values = processor(image, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)

generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
```
